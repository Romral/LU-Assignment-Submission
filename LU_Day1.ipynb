# Line Plot

Displays information as a series of datapoints called markers, connected by straight line segments

```python
x = np.arange(0,11)
y = np.arange(10,21)
print('x:',x)
print('y:',y)
```

```python
plt.plot(x,y, 'go--', linewidth=2, markersize=12)
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y)
```

```python
x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Waveform")

plt.plot(x,y,color = 'b')
plt.show()#Not mandatory but the differnce is no matplotlib line before graph 
```

```python
x = np.arange(0,10)
y = x*x
plt.plot(x,y,'r*', linestyle = 'dashed', linewidth = 2, markersize = 12)
plt.title("2D-Diagram",color='b')
plt.xlabel("X-Axis",color='b')
plt.ylabel("Y-Axis",color='b')
```

```python
y = 3*x + 10
plt.plot(x,y,'ro--')
plt.title("2D Diagram")
plt.xlabel("X-Axis",color='b')
plt.ylabel("Y-Axis",color='b')
```

```python
#Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = [1,2,3,4,5,6,7]
y = [160,150,140,145,175,165,180]
plt.plot(x,y)
plt.title("Sales vs Week Days")
plt.xlabel("Days of the Week")
plt.ylabel("Sales")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70, 90, 160, 150, 140, 145, 175]
plt.figure(figsize=(6,5), dpi=150)
plt.plot(days, sales_1)
plt.plot(days, sales_2)
plt.title("Sales vs Days of week for Company 1 & 2")
plt.xlabel("days of the week")
plt.ylabel("sales")
plt.grid()
plt.show()
```
