```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style = 'darkgrid')
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.describe().transpose()
```

```python
sns.relplot(x = 'timepoint', y='signal', hue='region', data = fmri, style = 'event')
```

```python
sns.lineplot(x='timepoint', y='signal', style = 'event', hue='region', data = fmri)
```

```python
sns.scatterplot(x='timepoint', y='signal', data=fmri )
```

```python
sns.catplot(x = 'subject', y= 'signal', data = fmri, hue = 'region', jitter = False)
```

```python
sns.catplot(x = 'subject', y= 'signal', data = fmri, hue = 'region', kind = 'strip')
```

```python
sns.catplot(x = 'subject', y= 'signal', data = fmri, hue = 'region', kind = 'boxen')
```

```python
sns.catplot(x = 'subject', y= 'signal', data = fmri, hue = 'region', kind = 'box')
```

```python
sns.catplot(y = 'subject', x = 'signal', data = fmri, hue = 'region', kind = 'violin', split = True)
```
